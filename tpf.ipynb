# Estimación

## Modelo 1 usando LM (Linear Model)

```r
mod1 = lm(new_base, formula = l_total_ingresos_1930 ~ l_gasto_ssa_qcm_2290 + l_total_activo_1080 )
summary(mod1)
```

## Modelo 2 PLM (Panel Data Fixed effects)

```r
library(plm)
mod2 = plm(new_base, formula = l_total_ingresos_1930 ~ l_gasto_ssa_qcm_2290 + l_total_activo_1080,
           model = "within",
           index = c("codigo_opera_parroquia", "anio_fiscal"))
summary(mod2)
```

## Modelo 3.1 usando librería prodest | levinsohn_petrin()

```r
mod3 = estprod::levinsohn_petrin(new_base,
                            formula = l_total_ingresos_1930 ~ l_gasto_ssa_qcm_2290 | l_total_activo_1080 | l_proxy_materiales, 
                            id = "codigo_opera_parroquia", time = "anio_fiscal", reps = 20)
summary(mod3)
```

## Modelo 3.2 Usando librería estprod | prodestLP()

```r
mod4 = prodestLP(new_base$l_total_ingresos_1930,
                 fX = new_base$l_gasto_ssa_qcm_2290,
                 sX =new_base$l_total_activo_1080, 
                 pX = new_base$l_proxy_materiales,
                 #new_base$l_gto_ivt_exploracion_otr_740,
                 idvar = new_base$codigo_opera_parroquia,
                 timevar = new_base$anio_fiscal,
                 opt='optim',
                 exit = F,
                 theta0 = NULL,
                 cluster = NULL,
                 tol = 1e-100)
summary(mod4)
```

```r
nivel_significancia = data.frame(mod4@Estimates$pars / mod4@Estimates$std.errors)
colnames(nivel_significancia)[1] = "val_significancia"
nivel_significancia$signif = if_else(nivel_significancia$val_significancia>=1.96, "significativo","")
nivel_significancia
```
