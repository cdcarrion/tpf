## Descripción de las variables utilizadas

total_ingresos_1930: Total de ingresos
costo_ssa_qcm_2280: Costo SUELDOS, SALARIOS Y DEMÁS REMUNERACIONES QUE CONSTITUYEN MATERIA GRAVADA DEL IESS
gasto_ssa_qcm_2290: Gasto SUELDOS, SALARIOS Y DEMÁS REMUNERACIONES QUE CONSTITUYEN MATERIA GRAVADA DEL IESS
total_activo_1080: Total del activo
costo_ivi_ppr_2240: Inventario inicial de productos en proceso
gto_ivt_exploracion_otr_740: Gastos de investigación exploración y similares

```r
new_base = dataSRI %>% 
  mutate(l_total_ingresos_1930 = log(total_ingresos_1930 + 1), 
         l_costo_ssa_qcm_2280 = log(costo_ssa_qcm_2280 + 1),
         l_gasto_ssa_qcm_2290 = log(gasto_ssa_qcm_2290 + 1),
         l_total_activo_1080  = log(total_activo_1080 + 1),
         l_costo_ivi_ppr_2240 = log(costo_ivi_ppr_2240 + 1)) %>%
  #Y - free: 2280 2290  - Kapital - Inversion
  select(l_total_ingresos_1930, 
         l_costo_ssa_qcm_2280, 
         l_gasto_ssa_qcm_2290,
         l_total_activo_1080, 
         l_costo_ivi_ppr_2240,
         codigo_opera_parroquia,
         anio_fiscal) %>% 
  arrange(codigo_opera_parroquia, anio_fiscal) #para que corra el codigo

summary(new_base)
```

```r
install.packages("prodest", repos='http://cran.us.r-project.org')
install.packages("estprod", repos='http://cran.us.r-project.org')
```

## Modelo  1 usando librería prodest | levinsohn_petrin()

```r
library(prodest)
library(estprod)

mod1 = estprod::levinsohn_petrin(new_base,
                            formula = l_total_ingresos_1930 ~ l_costo_ssa_qcm_2280 + l_gasto_ssa_qcm_2290 | l_total_activo_1080 | l_costo_ivi_ppr_2240, 
                            id = "codigo_opera_parroquia", time = "anio_fiscal", reps = 20)

summary(mod1)
```

## Modelo 2 Usando librería estprod | prodestLP()

```r
mod2 = prodestLP(new_base$l_total_ingresos_1930,
                 fX = cbind(new_base$l_costo_ssa_qcm_2280, new_base$l_gasto_ssa_qcm_2290),
                 sX =new_base$l_total_activo_1080, 
                 pX = new_base$l_costo_ivi_ppr_2240,
                 #new_base$l_gto_ivt_exploracion_otr_740,
                 idvar = new_base$codigo_opera_parroquia,
                 timevar = new_base$anio_fiscal,
                 opt='optim',
                 exit = F,
                 theta0 = NULL,
                 cluster = NULL,
                 tol = 1e-100)
summary(mod2)
```

```r
nivel_significancia = data.frame(mod2@Estimates$pars / mod2@Estimates$std.errors)
colnames(nivel_significancia)[1] = "val_significancia"
nivel_significancia$signif = if_else(nivel_significancia$val_significancia>=1.96, "significativo","")
nivel_significancia
```
